```python
import cv2
import matplotlib.pyplot as plt

def balance_de_blancos(imagen):
    # Cargar la imagen
    img = cv2.imread(imagen)

    # Aplicar el balance de blancos utilizando el algoritmo de Gray World
    resultado = cv2.xphoto.createSimpleWB()
    img_balanceada = resultado.balanceWhite(img)

    # Mostrar la imagen original y la imagen con balance de blancos
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title('Imagen Original')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    
    plt.subplot(1, 2, 2)
    plt.title('Imagen con Balance de Blancos')
    plt.imshow(cv2.cvtColor(img_balanceada, cv2.COLOR_BGR2RGB))
    
    # Mostrar las imágenes en una ventana externa
    plt.show()

# Ruta de la imagen que quieres procesar
ruta_imagen = 'stone.jpg'

# Llamar a la función de balance de blancos
balance_de_blancos(ruta_imagen)
```
